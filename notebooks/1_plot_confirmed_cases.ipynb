```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, timedelta
```

```python
df1 = pd.read_csv("../data/processed/enviroment_provinces_Mainland China.csv")
df2 = pd.read_csv("../data/processed/enviroment_provinces_Italy.csv")
```

```python
fig = plt.figure(figsize=(15,5))
countries = ['Mainland China', 'Italy']
for i, df in enumerate([df1, df2]):
    plt.subplot(1,2,i+1)
    for region in df['Province/State'].unique():
        tmp = df[df['Province/State'] == region]
        plt.semilogy(tmp['date'], np.array(tmp['Confirmed']).astype('float64'))
    didx = 15
    tidx = np.arange(75, len(tmp), didx)
    idx = tmp['date'].index[tidx]
    plt.xticks(tmp['date'][idx], rotation=90)
    plt.title(countries[i])
    plt.ylabel("Confirmed COVID-19 cases")
    plt.grid(True)
    plt.ylim([1.e0, 1.e5])
    
plt.tight_layout()
plt.savefig("../figures/confirmed_cases_Italy_and_China.png", dpi=200, bbox_inches='tight')
```
